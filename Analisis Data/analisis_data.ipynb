# ANALISIS DATA DENGAN VISUALISASI

Pada tahap ini, kita akan melakukan eksplorasi data diabetes menggunakan visualisasi. Analisis dilakukan untuk memahami distribusi data dan hubungan antar variabel sebelum proses klasifikasi.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
data = pd.read_csv('diabetes.csv')

# Cek 5 data teratas untuk memastikan
print(data.head())
```

# 2. ANALISIS UNIVARIATE

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# sns style
sns.set(style="whitegrid")

# Histogram distribusi semua variabel
data.hist(bins=15, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle("Distribusi Setiap Variabel", fontsize=16)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', hue='Outcome', data=data, palette='viridis', legend=False)
plt.title('Distribusi Target (Outcome)')
plt.xlabel('Outcome (0 = Negatif, 1 = Positif)')
plt.ylabel('Jumlah')
plt.show()
```

# 3. ANALISIS BIVARIATE

```python
plt.figure(figsize=(15, 10))

# Loop setiap kolom untuk dibandingkan dengan Outcome
for i, col in enumerate(['Glucose', 'BMI', 'Age', 'BloodPressure']):
    plt.subplot(2, 2, i + 1)
    
    sns.boxplot(
        x='Outcome',
        y=col,
        hue='Outcome',
        data=data,
        palette='coolwarm',
        legend=False  
    )
    
    plt.title(f'{col} vs Outcome')

plt.tight_layout()
plt.show()
```

# 4. ANALISIS MULTIVARIATE

```python
plt.figure(figsize=(10, 8))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title("Matriks Korelasi antar Variabel")
plt.show()
```

# Kesimpulan

- Beberapa variabel seperti `Glucose`, `BMI`, dan `Age` memiliki perbedaan yang jelas antara pasien diabetes dan non-diabetes.
- Korelasi antar fitur menunjukkan beberapa fitur saling berkaitan, seperti Glucose dan Outcome.
- Visualisasi ini membantu dalam memilih fitur yang relevan untuk proses klasifikasi selanjutnya.
